```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
boston = datasets.load_boston()
```

```python
boston.data.shape
```

```python
lstat = boston.data[:, np.newaxis, 12]
rm = boston.data[:, np.newaxis, 5]
```

```python
target = boston.target
```

```python
target
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.scatter(lstat, rm, target, color='red')
plt.show()
```

```python
#data training pada regresi ini adalah kombinasi dari atribut lstat dan rm
data_training = np.c_[lstat,rm]

regr= linear_model.LinearRegression()
regr.fit(data_training, target)
```

```python
target_predict = regr.predict(data_training)
rmse=(np.sqrt(mean_squared_error(target, target_predict)))

print (rmse)
```
