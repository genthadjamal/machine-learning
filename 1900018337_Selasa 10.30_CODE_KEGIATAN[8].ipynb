```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(random_state=0, n_samples=1000)
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X[:,0],X[:,1],'ro')
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="ward", affinity='euclidean')
```

```python
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clustering.labels_, marker="o")
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="average", affinity='euclidean')
```

```python
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clustering.labels_, marker="o")
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="complete", affinity='euclidean')
```

```python
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clustering.labels_, marker="o")
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="average", affinity='manhattan')
```

```python
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clustering.labels_, marker="o")
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="complete", affinity='manhattan')
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clustering.labels_, marker="o")
plt.show()
```
